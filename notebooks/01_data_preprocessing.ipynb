```python
import pandas as pd
from pathlib import Path

cic_dir = Path(r"C:\ShieldNet-AI\cic")

cic_files = list(cic_dir.glob("*.csv"))
print(f"Found {len(cic_files)} CIC files")

dfs = []
for file in cic_files:
    df = pd.read_csv(file)
    print(f"{file.name}: {df.shape}")
    dfs.append(df)
```

```python
all_columns = set()

for df in dfs:
    all_columns.update(df.columns)

sorted(all_columns)
```

```python
CIC_TO_PCAP_MAP = {
    "Flow Duration": "Flow Duration",
    "Total Fwd Packets": "Total Fwd Packets",
    "Total Backward Packets": "Total Bwd Packets",
    "Fwd Packets Length Total": "Total Length Fwd",
    "Fwd Packet Length Max": "Fwd Packet Max",
    "Idle Mean": "Idle Mean",
    "Label": "Label"
}
```

```python
required_cic_cols = CIC_TO_PCAP_MAP.keys()

for col in required_cic_cols:
    if col not in all_columns:
        raise ValueError(f"❌ CIC column missing for mapping: {col}")

print(" All required CIC columns found")
```

```python
sorted(all_columns)
```

```python
CIC_TO_PCAP_MAP = {
    "Flow Duration": "Flow Duration",
    "Total Fwd Packets": "Total Fwd Packets",
    "Total Backward Packets": "Total Bwd Packets",
    "Total Length of Fwd Packets": "Total Length Fwd",
    "Fwd Packet Length Max": "Fwd Packet Max",
    "Idle Mean": "Idle Mean",
    "Label": "Label"
}
```

```python
required_cic_cols = CIC_TO_PCAP_MAP.keys()

missing = [c for c in required_cic_cols if c not in all_columns]
missing
```

```python
sorted(all_columns)
```

```python
missing == []
```

```python
normalized_dfs = []

for df in dfs:
    df_reduced = df[list(CIC_TO_PCAP_MAP.keys())].rename(columns=CIC_TO_PCAP_MAP)
    normalized_dfs.append(df_reduced)

cic_final = pd.concat(normalized_dfs, ignore_index=True)

print("Final CIC shape:", cic_final.shape)
cic_final.head()
```

```python
cic_final = cic_final.astype({
    "Flow Duration": "int64",
    "Total Fwd Packets": "int64",
    "Total Bwd Packets": "int64",
    "Total Length Fwd": "int64",
    "Fwd Packet Max": "int64",
    "Idle Mean": "float64",
    "Label": "object"
})

cic_final.dtypes
```

```python
# 1. No missing values
print(cic_final.isna().sum())

# 2. Columns must match exactly
print(cic_final.columns.tolist())
```

```python
cic_final.to_csv(r"C:\ShieldNet-AI\data/raw/cic_aligned_training.csv",
    index=False
)
```
